```python
import warnings
warnings.filterwarnings("ignore")
from nilmtk.api import API
from nilmtk.disaggregate import Mean
```

```python
from nilmtk_contrib.disaggregate import DAE,Seq2Point, Seq2Seq, RNN, WindowGRU
```

```python
ukdale = {
  'power': {
    'mains': ['apparent','active'],
    'appliance': ['apparent','active']
  },
  'sample_rate': 300,
  #'appliances': ['kettle', 'fridge freezer', 'washer dryer', 'dish washer', 'microwave', 'television'],
  'appliances': ['fridge freezer'],
  #'pre-trained': {'Seq2Point': './Weights/seq2point-temp-weights-65788.h5'},
  'methods': {
    
      #'WindowGRU':WindowGRU({'n_epochs':1,'batch_size':32}),
      #'RNN':RNN({'n_epochs':1,'batch_size':32}),
      #'DAE':DAE({'n_epochs':1,'batch_size':32}),
      'Seq2Point':Seq2Point({'n_epochs':30,'batch_size':32, 'sequence_length': 99}),
      #'Seq2Seq':Seq2Seq({'n_epochs':1,'batch_size':32}),

      'Mean': Mean({}),
          
  },
   'train': {    
    'datasets': {
            'UK-DALE': {
                'path': './Dataset/ukdale.h5',
                #'path': './Weights/boiler_seq2point-temp-weights-45573.h5',
                'buildings': {
                1: {
                    'start_time': '2014-01-01',
                    'end_time': '2014-03-31'
                },
#                 56: {
#             'start_time': '2015-01-28 00:00:00',
#             'end_time': '2015-01-30 00:00:00'
#           },
            }
                                
            }
            }
    },
    'test': {
    'datasets': {
        'UK-DALE': {
            'path': './Dataset/ukdale.h5',
            'buildings': {
                1: {
                    'start_time': '2014-05-05',
                    'end_time': '2014-05-12'
                },
            }
        }
},
        'metrics':['mae', 'rmse']
}
}
```

The NILMTK API is used to call different disaggregation methods including Seq2Point, which is one of the methods defined in the 'methods' dictionary with specific parameters such as 'n_epochs' and 'batch_size'. The API is also specifying the train and test datasets, along with the evaluation metrics. This code is setting up the overall experiment to evaluate the performance of the different disaggregation methods on the specified dataset and appliance. The API will take care of loading the dataset, splitting it into train and test sets based on the specified dates, and training and evaluating each of the specified models. The output will include evaluation metrics such as mean absolute error (MAE) for each model on the test set.


```python
api = API(ukdale)
```

## Possible code to use pre-trained data

```python
test = {
  'power': {
    'mains': ['apparent','active'],
    'appliance': ['apparent','active']
  },
  'sample_rate': 60,
  #'appliances': ['television','kettle', 'boiler', 'freezer', 'washing machine'],
  'appliances': ['boiler'],
  'methods': {
    
      #'WindowGRU':WindowGRU({'n_epochs':1,'batch_size':32}),
      #'RNN':RNN({'n_epochs':1,'batch_size':32}),
      #'DAE':DAE({'n_epochs':1,'batch_size':32}),
      'Seq2Point':Seq2Point({'n_epochs':5,'batch_size':32,'pretrained_path': './seq2point-temp-weights-52039.h5' }),
      #'Seq2Seq':Seq2Seq({'n_epochs':1,'batch_size':32}),

      'Mean': Mean({}),
          
  },
   'train': {    
    'datasets': {
            'UK-DALE': {
                'path': './Dataset/ukdale.h5',
                'buildings': {
                4: {
                    'start_time': '2013-04-01 00:00:00',
                    'end_time': '2013-04-30 00:00:00'
                },
#                 56: {
#             'start_time': '2015-01-28',
#             'end_time': '2015-01-30'
#           },
            }
                                
            }
            }
    },
    'test': {
    'datasets': {
        'UK-DALE': {
            'path': './Dataset/ukdale.h5',
            'buildings': {
                4: {
                    'start_time': '2013-06-02 00:00:00',
                    'end_time': '2013-06-02 12:00:00'
                },
            }
        }
},
        'metrics':['mae', 'rmse']
}
}
```

It tells the Seq2Point model to load the weights stored in the file at that path before starting the training process. This means that the model will start with the same weights that were learned during a previous training process, rather than starting with random weights.

So, when you call the API method and pass in your modified ukdale dictionary as a parameter, the Seq2Point model will be initialized with the pre-trained weights stored in the 'pre_trained_path' file, and then it will be fine-tuned using the data and settings specified in the ukdale dictionary. The resulting trained model will then be used to make predictions on the test data, and the evaluation metrics will be computed based on these predictions

```python
pretrained = API(test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define appliance 
appliance = 'boiler'

# Convert the index to a datetime object for ground truth
df_test = pretrained.gt_overall.copy()
df_test.index = pd.to_datetime(df_test.index)

# Convert the predicted values dictionary to a dataframe
pred_df_test= pd.DataFrame(pretrained.pred_overall['Seq2Point'])
pred_df_test.index = pd.to_datetime(pred_df_test.index)

# Plot the ground truth and predicted values in the same graph
plt.plot(df_test[appliance], label='Ground Truth')
plt.plot(pred_df_test[appliance], label='Predicted')
plt.xlabel('Time')
plt.ylabel('Power (Watts)')
plt.title(appliance.capitalize() + ' Power Consumption')
plt.legend()
plt.show()
```
